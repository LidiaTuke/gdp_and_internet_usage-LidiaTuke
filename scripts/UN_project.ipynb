```python
gdp_and_internet_use = pd.merge(gdp_df, internet_df, left_on= ['Country', 'Year'], right_on= ['Country','Year'], how='inner')
gdp_and_internet_use
```

```python
gdp_and_internet_2014 = gdp_and_internet_use.loc[gdp_and_internet_use['Year'] == 2014]
gdp_and_internet_2014
```

```python
gdp_and_internet_2014= gdp_and_internet_2014.sort_values(by = 'Internet_Users_Pct', ascending = False).head(3)
gdp_and_internet_2014
```

```python
top_country = gdp_and_internet_use.loc[gdp_and_internet_use['Country'].isin (['Iceland', 'Bermuda', 'Norway'])]
top_country
```

```python
top_internet_user_country = sns.FacetGrid(top_country, col='Country')
top_internet_user_country.map(sns.scatterplot,'Year', 'GDP_Per_Capita');
```
